```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
len(soup.find_all('table'))
```

```python
s = '''<a href="/wiki/SpaceX" title="SpaceX">SpaceX</a>'''
re.sub("<a.*>", "", s)
```

```python
def clean(s):
    s = s.replace("<tr>", "").replace("</tr>", "")
    s = s.replace("<th>","").replace("</th>",";")
    s = s.replace("<td>", "").replace("</td>", "")
    
    return s
    
for row in soup.find_all('table')[1]:
    print clean(str(row))
    print "----------"
    
```
