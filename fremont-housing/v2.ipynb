# remove columns

```python
use_columns = ["zipcode", "area", "bet", "bath", "sqft", "Zestimate", "type", "built", "year1", "price1", "year2", "price2", "year3", "price3", "year4", "price4", "year5", "price5"]
data = data.loc[:, use_columns]
sns.barplot(y=data.isnull().sum().index, x=data.isnull().sum().values*1.0 / data.shape[0])
```

# manipulate historical data

```python
static = data.loc[:, ["zipcode", "area", "bet", "bath", "sqft", "type", "built"]]
print np.array(static).shape
```

```python
static_expand = np.repeat(np.array(static), 6, axis=0)
print static_expand.shape
```

```python
hist = data.loc[:, ["type", "Zestimate", "year1", "price1", "year2", "price2", "year3", "price3", "year4", "price4", "year5", "price5"]]
hist.rename(columns = {'type':'year0', 'Zestimate':'price0'}, inplace=True)
hist.year0 = 2017
hist.head()
```

```python
hist_np = np.array(hist)
hist_long = []
for i in xrange(hist_np.shape[0]):
    for j in xrange(6):
        hist_long.append([hist_np[i,j*2], hist_np[i,j*2+1], i])

hist_df = np.array(hist_long)
print hist_df.shape
```

```python
hist_full = pd.DataFrame(np.hstack((static_expand, hist_df)), columns=["zipcode", "area", "bet", "bath", "sqft", "type", "built","year","price", "house_id"])
print hist_full.shape
hist_full.head()
```

```python
def cleanYear(s):
    if s=='2017':
        return 0
    try:
        r = (datetime.strptime("2017-07-16", '%Y-%m-%d') - datetime.strptime(s, '%Y-%m-%d')).days
        return r
    except:
        return np.nan
```

```python
hist_full.loc[:, "year"] = [cleanYear(s) for s in hist_full.year]
```
